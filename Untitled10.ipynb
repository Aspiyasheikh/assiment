Q6. Explain the concept of multicollinearity in multiple linear regression.How can youdetect and address this issue?

Ans:  What is Multicollinearity?
Multicollinearity occurs when two or more independent variables in a regression model are highly correlated â€” meaning they provide similar information about the dependent variable.

ğŸ” Why is it a problem?
Makes it difficult to estimate the effect of each independent variable accurately

Leads to unstable coefficients (they may change drastically with small data changes)

Reduces the interpretability of the model

How to Detect Multicollinearity:
Method	Description
Correlation Matrix	Check if any predictors have high correlation (r > 0.8 or r < -0.8)
Variance Inflation Factor (VIF)	VIF > 5 or 10 indicates high multicollinearity
Condition Number	Large values (> 30) suggest multicollinearity

Q8. What are the advantages nd disadvantages of polynomial regression compared to linear regression? in what situations would you prefer to use polynomial regression?

Ans: Advantages of Polynomial Regression:
Captures non-linear relationships: It can model complex curves that linear regression cannot.

Better fit for curved data: If the data forms a curved pattern, polynomial regression provides a closer fit.

Flexible: You can increase the degree to improve accuracy (with caution).

âŒ Disadvantages of Polynomial Regression:
Overfitting: Higher-degree polynomials may fit the training data too closely and perform poorly on new data.

Computational cost: More complex models require more computation and time.

Less interpretable: As the degree increases, it becomes harder to understand the relationship.

Sensitive to outliers: A few extreme points can distort the curve significantly.

ğŸ“Œ When to use Polynomial Regression:
When the data is non-linear and a straight line doesnâ€™t fit well.

When visualization or residuals suggest a curved pattern.

When underfitting occurs with linear regression and performance improves with added degrees.
